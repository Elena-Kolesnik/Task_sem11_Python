```python
import numpy as np
from sympy import *
from sympy.plotting import plot
x=Symbol('x')
f = -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30

def func(x):
    return -12 * x**4 * np.sin(np.cos(x)) - 18 * x**3 + 5 * x**2 + 10 * x - 30
```

Количество корней и экстремумов (вершин) бесконечно, следовательно вычисление конкретных значений актуально в определенном интервале значений х (в задании не определено).
```python
def bisection(f, lhs, rhs, eps=1e-6):

  mid = (lhs + rhs) / 2

  while abs(lhs - rhs) > eps:
    mid = (lhs + rhs) / 2

    if f(lhs) * f(mid) > 0: # f(lhs) > 0 and f(mid) > 0 or f(lhs) < 0 and f(mid) < 0
      lhs = mid
    else:
      rhs = mid
      
  return mid
```
